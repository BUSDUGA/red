```python
import os
os.environ["HF_HUB_DISABLE_XET"] = "1"
os.environ["HF_HUB_DOWNLOAD_TIMEOUT"] = "300"
os.environ["HF_HUB_ETAG_TIMEOUT"] = "300"

import json
import gc
import time
import pickle
import random
from pathlib import Path

import numpy as np
import torch
import torch.nn.functional as F

from tqdm.auto import tqdm
from scipy.sparse import hstack
from huggingface_hub import snapshot_download

from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    AutoModelForSequenceClassification,
    BitsAndBytesConfig,
)

from peft import (
    LoraConfig,
    get_peft_model,
    prepare_model_for_kbit_training,
    PeftModel,
)

SEED = 42

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)


DEVICE = "cuda"

DATA = Path("/content/task_data/ml-olympiad-red-task/data")
METRICS = Path("/content/task_data/ml-olympiad-red-task/metrics")

def read_jsonl(path):
    a = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            a.append(json.loads(line))
    return a
kid_adult = read_jsonl(DATA / "kid_adult.jsonl")
good_bad = read_jsonl(DATA / "good_bad.jsonl")
test_style = read_jsonl(DATA / "public_test_style.jsonl")
test_quality = read_jsonl(DATA / "public_test_quality.jsonl")

print("kid_adult:", len(kid_adult))
print("good_bad:", len(good_bad))
print("test_style:", len(test_style))
print("test_quality:", len(test_quality))

MODEL_NAME = "Qwen/Qwen3-4B-Instruct-2507"
MODEL_ID = "/content/qwen_model"

t = time.time()

snapshot_download(
    repo_id=MODEL_NAME,
    local_dir=MODEL_ID,
    max_workers=1,
    allow_patterns=[
        "*.json",
        "*.safetensors",
        "tokenizer*",
        "*.model",
        "*.py",
        "*.txt",
    ],
)

MAX_LEN = 512
MAX_NEW = 128
LR = 2e-4
BETA = 0.1
GAMMA = 0.1
```

```python
tok = AutoTokenizer.from_pretrained(
    MODEL_ID,
    trust_remote_code=True,
    use_fast=True,
)

if tok.pad_token is None:
    tok.pad_token = tok.eos_token

tok.padding_side = "left"

bnb = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

lora_lm = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.05,
    target_modules=["q_proj", "v_proj"],
    task_type="CAUSAL_LM",
)

lora_rm = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.05,
    target_modules=["q_proj", "v_proj"],
    task_type="SEQ_CLS",
    modules_to_save=["score"],
)

def chat(prompt):
    return tok.apply_chat_template(
        [{"role": "user", "content": prompt}],
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=False,
    )

def new_lm(adapter=None, train=True):

    m = AutoModelForCausalLM.from_pretrained(
        MODEL_ID,
        quantization_config=bnb,
        device_map="auto",
        trust_remote_code=True,
    )

    m.config.use_cache = False
    m = prepare_model_for_kbit_training(m)

    if adapter is None:
        m = get_peft_model(m, lora_lm)
    else:
        m = PeftModel.from_pretrained(m, adapter, is_trainable=train)

    if train:
        m.train()
    else:
        m.eval()
        for p in m.parameters():
            p.requires_grad = False

    return m

def new_rm():

    m = AutoModelForSequenceClassification.from_pretrained(
        MODEL_ID,
        num_labels=1,
        quantization_config=bnb,
        device_map="auto",
        trust_remote_code=True,
    )

    m.config.pad_token_id = tok.pad_token_id
    m = prepare_model_for_kbit_training(m)
    m = get_peft_model(m, lora_rm)
    m.train()

    return m

def logp(model, prompt, answer):
    start = chat(prompt)
    text = start + answer + tok.eos_token

    a = tok(start, return_tensors="pt").input_ids.shape[1]

    x = tok(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=MAX_LEN,
    ).to(DEVICE)

    out = model(**x).logits[:, :-1]
    y = x.input_ids[:, 1:]

    lp = F.log_softmax(out, dim=-1)
    lp = lp.gather(2, y[:, :, None]).squeeze(2)

    mask = torch.zeros_like(y, dtype=torch.bool)
    mask[:, a - 1:] = True

    return lp[mask].mean()

def rm_score(model, prompt, answer):
    text = chat(prompt) + answer

    x = tok(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=MAX_LEN,
    ).to(DEVICE)

    return model(**x).logits.view(-1)[0]

def gen_answers(model, tests):
    model.eval()
    ans = []

    for x in tqdm(tests, desc="generate"):
        q = tok(chat(x["prompt"]), return_tensors="pt").to(DEVICE)

        with torch.no_grad():
            out = model.generate(
                **q,
                max_new_tokens=MAX_NEW,
                do_sample=False,
                pad_token_id=tok.eos_token_id,
            )

        text = tok.decode(
            out[0][q.input_ids.shape[1]:],
            skip_special_tokens=True,
        ).strip()

        ans.append(text)

    model.train()
    return ans

style = pickle.load(open(METRICS / "style_clf.pkl", "rb"))

def p_simple(texts):
    v1, v2 = style["vecs"]
    X = hstack([v1.transform(texts), v2.transform(texts)])
    return float(style["clf"].predict_proba(X)[:, 1].mean())

def quality_acc(model, data):
    ok = 0
    model.eval()

    for x in tqdm(data, desc="quality acc"):
        with torch.no_grad():
            good = logp(model, x["instruction"], x["chosen"])
            bad = logp(model, x["instruction"], x["rejected"])

        if good > bad:
            ok += 1

    model.train()
    return ok / len(data)

def rm_acc(model, data):
    ok = 0
    model.eval()

    for x in tqdm(data, desc="rm acc"):
        with torch.no_grad():
            good = rm_score(model, x["instruction"], x["chosen"])
            bad = rm_score(model, x["instruction"], x["rejected"])

        if good > bad:
            ok += 1

    model.train()
    return ok / len(data)


def clean():
    gc.collect()
    torch.cuda.empty_cache()

```

```python
model = new_lm()

opt = torch.optim.AdamW(
    [p for p in model.parameters() if p.requires_grad],
    lr=LR,
)

print("Задача 1: SFT")

for x in tqdm(kid_adult, desc="SFT"):
    loss = -logp(model, x["prompt"], x["kid"])

    loss.backward()
    opt.step()
    opt.zero_grad()

model.save_pretrained("sft")

answers = gen_answers(model, test_style)

m1 = p_simple(answers)

print("Задача 1")
print("P_simple =", m1)

del model
clean()
```

```python
ref = new_lm(adapter=None, train=False)

ref_style = []

for x in tqdm(kid_adult, desc="ref style"):
    with torch.no_grad():
        good = logp(ref, x["prompt"], x["kid"])
        bad = logp(ref, x["prompt"], x["adult"])

    ref_style.append(float(good - bad))

del ref
clean()

model = new_lm(adapter=None, train=True)

opt = torch.optim.AdamW(
    [p for p in model.parameters() if p.requires_grad],
    lr=LR,
)

for x, r in tqdm(list(zip(kid_adult, ref_style)), desc="DPO style"):
    good = logp(model, x["prompt"], x["kid"])
    bad = logp(model, x["prompt"], x["adult"])

    diff = good - bad
    ref_diff = torch.tensor(r, device=DEVICE)

    loss = -F.logsigmoid(BETA * (diff - ref_diff))

    loss.backward()
    opt.step()
    opt.zero_grad()

model.save_pretrained("/content/dpo_style")

answers = gen_answers(model, test_style)

m2 = p_simple(answers)

print("Задача 2")
print("P_simple =", m2)


del model
clean()
```
